```python


# Create a figure with subplots
fig, axes = plt.subplots(5, 2, figsize=(12, 16))

# Plotting the first distribution plot
ax1 = axes[0, 0]
sns.distplot(df['Q'], bins=15, ax=ax1)
ax1.set_title("Distribution of Q")

# Plotting the second distribution plot
ax2 = axes[0, 1]
sns.distplot(df['U'], bins=15, ax=ax2)
ax2.set_title("Distribution of U")

# Plotting the third distribution plot
ax3 = axes[1, 0]
sns.distplot(df['W'], bins=10, ax=ax3)
ax3.set_title("Distribution of W")

# Plotting the fourth distribution plot
ax4 = axes[1, 1]
sns.distplot(df['S'], bins=10, ax=ax4)
ax4.set_title("Distribution of S")

# Plotting the fifth distribution plot
ax5 = axes[2, 0]
sns.distplot(df['H'], bins=10, ax=ax5)
ax5.set_title("Distribution of H")

# Plotting the sixth distribution plot
ax6 = axes[2, 1]
sns.distplot(df['R'], bins=10, ax=ax6)
ax6.set_title("Distribution of R")

# Plotting the seventh distribution plot
ax7 = axes[3, 0]
sns.distplot(df['D50'], bins=10, ax=ax7)
ax7.set_title("Distribution of D50")

# Plotting the eighth distribution plot
ax8 = axes[3, 1]
sns.distplot(df['D84'], bins=10, ax=ax8)
ax8.set_title("Distribution of D84")
ax8 = axes[4, 0]
sns.distplot(df['qs'], bins=10, ax=ax8)
ax8.set_title("Distribution of qs")

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
size_95_percent = np.percentile(df['qs'], 5)

# Print the size or use it for further analysis
print("Size of 'qs' where 95% of data points are smaller:", size_95_percent)
count_qs_greater_than_n = len(df[df['qs'] > 63000])
# Print the count or use it for further analysis
print("Number of data points where qs > n:", count_qs_greater_than_n)

df_filtered = df[(df['qs'] >= 0.05) & (df['qs'] <= 63000)]
df=df_filtered
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a correlation plot using seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)

# Set the title and display the plot
plt.title('Correlation Plot')
plt.show()
```
